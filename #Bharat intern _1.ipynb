```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\SRIHARI SEKHAPURAM\Documents\Dataset .csv")
print(df)
```

```python
cuisine_data = df["Cuisines"]
```

```python
cuisine_counts=cuisine_data.value_counts()
```

```python
top_cuisines=cuisine_counts.head(3)
```

```python
total_restaurants = len(df)
percentage= (top_cuisines/ total_restaurants)*100
```

```python
plt.bar(percentage.index,percentage.values)
plt.xlabel("Cuisines")
plt.ylabel("Percentage")
plt.title("top Three Companies in Data")
plt.show()
```

```python
plt.pie(top_cuisines,labels=top_cuisines.index,autopct='%1.1f%%')
plt.title('top three cuisines in data set')
plt.show()
```
