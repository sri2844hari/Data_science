```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:\\Users\\SRIHARI SEKHAPURAM\Documents\Dataset .csv")
```

```python
restarent_count = data.groupby('City')['Restaurant ID'].count()
```

```python
average_rating = data.groupby('City')['Aggregate rating'].mean()
```

```python
average_rating.plot(kind='bar',figsize=(26,8))
plt.xlabel("City")
plt.ylabel("averae rating")
plt.title("Average rating for restarent in eeach city")
plt.show()
```

```python
city_with_highest_average_ratings =average_rating.idxmax()
print("city with highest average rating: ",city_with_highest_average_ratings)
```
