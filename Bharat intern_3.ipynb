```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("C:\\Users\\SRIHARI SEKHAPURAM\Documents\Dataset .csv")
```

```python
price_range=dataset['Price range'].value_counts().sort_index()
```

```python
total_restarents = len(dataset)
percentage_perprice_range = (price_range/total_restarents)*100
```

```python
plt.bar(price_range.index,price_range.values)
plt.xlabel('price range')
plt.ylabel("count")
plt.title("Distribution of price range")
plt.show()
```
