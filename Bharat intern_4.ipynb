```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("C:\\Users\\SRIHARI SEKHAPURAM\Documents\Dataset .csv")
```

```python
total_restarents=len(dataset)
print(total_restarents)
```

```python
restarents_with_online_delivary=dataset['Has Online delivery'].value_counts().get('Yes',0)
Percentage_with_onilne_delivary=(restarents_with_online_delivary/total_restarents)*100
print("Percentage of restarents having online delivary: ",Percentage_with_onilne_delivary)
```

```python
average_rating_with_online_delivary=dataset[dataset['Has Online delivery']=='Yes']['Aggregate rating'].mean()
average_rating_without_online_delivary=dataset[dataset['Has Online delivery']=='No']['Aggregate rating'].mean()
```

```python
print("Average rating of restarent with online delivar:  ",average_rating_with_online_delivary)
print("Average rating of restarent without online delivar:  ",average_rating_without_online_delivary)
```

```python
labels=['With Online delivery','Without Online delivery']
average_rating=[average_rating_with_online_delivary,average_rating_without_online_delivary]
plt.bar(labels,average_rating)
plt.xlabel('OnlineDelivery')
plt.ylabel('Average Rating')
plt.title('comparsion of average rating')
plt.show()
```
